```python
import fitz
import time
import re
```

```python
def get_pic_from_pdf(path,save_path):
    file_name = path.split('\\')[-1].split(".")[0]
    doc = fitz.open(path)
    doc._getXrefLength()
    # 图片计数
    imgcount = 0
    lenXREF = doc._getXrefLength()
    t0 = time.time()
    # 使用正则表达式来查找图片
    checkXO = r"/Type(?= */XObject)" 
    checkIM = r"/Subtype(?= */Image)" 
    # 打印PDF的信息
    print("文件名:{}, 页数: {}, 对象: {}".format(path.split('/')[-1], len(doc), lenXREF - 1))

    # 遍历每一个对象
    for i in range(1, lenXREF):
        # 定义对象字符串
        # text = doc.getObjectString(i)
        text = doc._getXrefString(i)
        
        isXObject = re.search(checkXO, text)
        # 使用正则表达式查看是否是图片
        isImage = re.search(checkIM, text)
        # 如果不是对象也不是图片，则continue
        if not isXObject or not isImage:
            continue
        imgcount += 1
        # 根据索引生成图像
        pix = fitz.Pixmap(doc, i)
        # 根据pdf的路径生成图片的名称
        # new_name = path.replace('\\', '_') + "_img{}.png".format(imgcount)
        new_name =  file_name + "_{}.png".format(imgcount)
        new_name = new_name.replace(':', '')

        # 如果pix.n<5,可以直接存为PNG
        if pix.n < 5:
            pix.writePNG(os.path.join(save_path, new_name))
            # pix.writePNG(new_name)
        # 否则先转换CMYK
        else:
            pix0 = fitz.Pixmap(fitz.csRGB, pix)
            pix0.writePNG(os.path.join(save_path, new_name))
            pix0 = None
        # 释放资源
        pix = None
        t1 = time.time()
        print("运行时间:{}s".format(t1 - t0))
        print("提取了{}张图片".format(imgcount))
```
